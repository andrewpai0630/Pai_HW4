* Run basic calculations to obtain purchase count, avg. purchase price, etc. by gender


* For normalized purchasing, divide total purchase value by purchase count, by gender


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
male = purchase_data[purchase_data["Gender"] == "Male"]
avg_price_male = "${0:.2f}".format(float(male["Price"].mean()))
total_male = "{0:.2f}".format(float(male["Price"].sum()))
female = purchase_data[purchase_data["Gender"] == "Female"]
avg_price_female = "${0:.2f}".format(float(female["Price"].mean()))
total_female = "{0:.2f}".format(float(female["Price"].sum()))
nondc = purchase_data[purchase_data["Gender"] == "Other / Non-Disclosed"]
avg_price_nondc = "${0:.2f}".format(float(nondc["Price"].mean()))
total_nondc = "{0:.2f}".format(float(nondc["Price"].sum()))
normalized_male = "${0:.2f}".format(float(total_male) / len(male))
normalized_female = "${0:.2f}".format(float(total_female) / len(female))
normalized_nondc = "${0:.2f}".format(float(total_nondc) / len(nondc))
df_purchasing = pd.DataFrame({" ": ["Gender", "Female", "Male", "Other / Non-Disclosed"],
                              "Purchase Count": ["", len(female), len(male), len(nondc)],
                              "Average Purchase Price": ["", avg_price_female, avg_price_male, avg_price_nondc],
                              "Total Purchase Value": ["", "$"+total_female, "$"+total_male, "$"+total_nondc],
                              "Normalized Totals": ["", normalized_male, normalized_female, normalized_nondc]
                             })
df_purchasing.set_index(" ", inplace=True)
df_purchasing.head()
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Establish bins for ages
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
purchase_data["Age Group"] = pd.cut(purchase_data["Age"], age_bins, labels=group_names)
purchase_data_age_group = purchase_data.groupby("Age Group")
age_group_count = purchase_data_age_group.count()
age_group_pct = []
for i in range(len(group_names)):
     group_pct = (age_group_count["Age"][i] / age_group_count["Age"].sum())*100
     age_group_pct.append(group_pct)        
df_age_group = pd.DataFrame({"Percentage": age_group_pct,
                             "Total Count": age_group_count["Age"]
                           })
df_age_group["Percentage"] = df_age_group["Percentage"].map('{:,.2f}'.format)
df_age_group
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, etc. in the table below


* Calculate Normalized Purchasing


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
purchase_data["Age Group"] = pd.cut(purchase_data["Age"], age_bins, labels=group_names)
purchase_data_age_group = purchase_data.groupby("Age Group")
age_group_count = purchase_data_age_group.count()
total_value = []
average_price = []
normalized_average = [] 
for i in range(len(group_names)):
    total = purchase_data_age_group["Price"].sum()[i]
    total_value.append(total)
    average = purchase_data_age_group["Price"].mean()[i]
    average_price.append(average)
    normalized =  total / age_group_count["Age"][i]
    normalized_average.append(normalized)
df_age_analysis = pd.DataFrame({"Purchase Count": age_group_count["Age"],
                                "Average Purchase Price": average_price, 
                                "Total Purchase Value": total_value, 
                                "Normalized Average": normalized_average
})
df_age_analysis["Average Purchase Price"] = df_age_analysis["Average Purchase Price"].map('${:,.2f}'.format)
df_age_analysis["Total Purchase Value"] = df_age_analysis["Total Purchase Value"].map('${:,.2f}'.format)
df_age_analysis["Normalized Average"] = df_age_analysis["Normalized Average"].map('${:,.2f}'.format)
df_age_analysis
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
purchase_data.head()
purchase_data_SN_group = purchase_data.groupby("SN")
num_of_purchase = purchase_data["SN"].value_counts()
purchase_count = num_of_purchase.count()
average_price = []
total_price = []
for i in range(len(num_of_purchase)):
    average = purchase_data_SN_group["Price"].mean()[i]
    average_price.append(average)
    total = purchase_data_SN_group["Price"].mean()[i] * num_of_purchase[i]
    total_price.append(total)
df_top_spenders = pd.DataFrame({"Purchase Count": num_of_purchase, 
                                "Average Purchase Price": average_price,
                                "Total Purchase Value": total_price})
df_top_spenders = df_top_spenders.sort_values(by="Total Purchase Value", ascending=False)
df_top_spenders["Average Purchase Price"] = df_top_spenders["Average Purchase Price"].map('${:,.2f}'.format)
df_top_spenders["Total Purchase Value"] = df_top_spenders["Total Purchase Value"].map('${:,.2f}'.format)
df_top_spenders.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
items_df = purchase_data.loc[:, ["Item ID", "Item Name", "Price"]]
item_count = items_df["Item Name"].value_counts()
items_group = items_df.groupby(["Item ID", "Item Name"])
total_purchase = []
item_price = []
for i in range(len(items_df["Item Name"].value_counts())):
    price = items_df["Price"][i]
    item_price.append(price)
    total = items_df["Price"][i] * items_df["Item Name"].value_counts()[i]
    total_purchase.append(total)
popular_items = pd.DataFrame({
                              "Purchase Count": item_count,
                              "Item Price": item_price,
                              "Total Purchase Value": total_purchase})
popular_items["Item Price"] = popular_items["Item Price"].map('${:,.2f}'.format)
popular_items["Total Purchase Value"] = popular_items["Total Purchase Value"].map('${:,.2f}'.format)
popular_items
```
